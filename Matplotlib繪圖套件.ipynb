```python
import numpy as np
import matplotlib.pyplot as plt
def is_prime(n):
    if n<2:
        return False
    for i in range(2,int(np.sqrt(n))+1):
        if n%i==0:
            return False
    return True
def f(n):
    count=0
    for i in range(2,n+1):
        if is_prime(i):
            count+=1
    return count
```

```python
x=np.arange(2,101)
y=[f(n)for n in x]
fig,ax=plt.subplots()
ax.scatter(x,y)
ax.set_xlabel('n')
ax.set_ylabel('f(n)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f1(x, y):
    return np.sin(x) * np.sin(y)

def f2(x, y):
    return 1 / np.sqrt(x**2 + y**2 + 1)

def f3(x, y):
    return (x**2 + y) * np.exp(-x**2 - y**2)

fig = plt.figure(figsize=(10, 3))

ax1 = fig.add_subplot(131)
x = np.linspace(0, 2*np.pi, 200)
y = np.linspace(0, 2*np.pi, 200)
X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
levels = [-0.7, -0.3, 0, 0.3, 0.7]
CS = ax1.contour(X, Y, Z, levels=levels)
ax1.clabel(CS, inline=True, fontsize=10)

ax2 = fig.add_subplot(132, projection='3d')
X, Y = np.meshgrid(x, y)
Z = f2(X, Y)
ax2.plot_surface(X, Y, Z, cmap='viridis')

ax3 = fig.add_subplot(133, projection='3d')
X, Y = np.meshgrid(x, y)
Z = f3(X, Y)
ax3.contour(X, Y, Z, 6, cmap='viridis')
plt.subplots_adjust(wspace=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x=np.arange(-2,2,0.05)
y=np.arange(-2,2,0.05)
X,Y=np.meshgrid(x,y)
Z=(X+Y)*np.exp(-X**2-Y**2)
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,cmap='jet')
ax.view_init(elev=40,azim=160)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML


x = np.linspace(0,np.pi,256)
y = np.tan(np.sin(x))-(np.sin(np.tan(x)))
fig,ax=plt.subplots()
dot,=ax.plot([],[],'ro',markersize=10)
plt.close()

def init():
  ax.plot(x,y)
def animate(i):
  dot.set_data(x[i],y[i])

ani=FuncAnimation(fig=fig,
              func=animate,
              frames=240,
              init_func=init,
              interval=50)
HTML(ani.to_html5_video())
```
