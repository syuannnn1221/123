```python
#(5)
n=int(input("請輸入一個正整數(n): "))
while(n%2==0 or n<10 or n>16) :
  n=int(input("請重新輸入一個正整數(n): "))
mid=n//2+1
qq=0
a=mid
b=mid
c=0
d=n+1
for_end=n+1
print('    ',end='')
for i in range(1,for_end):
  print(f'{i:4d}',end='')
print()
for x in range(1,for_end):
  qq+=1;b+=1;a-=1
  if(x>mid):
    c+=1;d-=1
  print(f'{qq:4d}',end='')
  for y in range(1,for_end):
    if(x==mid or y==mid):
      print(' '*4,end='')
    elif(x<mid):
      if(y<a or y>b) :
       print(' '*4,end='')
      else:
       print(f'{x*y:4d}',end='')
    elif(x>mid):
      if(y<c or y>d):
        print(' '*4,end='')
      else:
        print(f'{x*y:4d}',end='')
  print()
```
