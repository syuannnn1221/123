7.設a=np.array([0,1,2,3,4,5])。在下列各題中，試說明b是a的一個view或是copy，並驗證您的判斷結果

```python
a=np.array([0,1,2,3,4,5])
#(a)
b=a[3:]
print(f'(a):{b}')
#(b)
b=a[::-1]
print(f'(b):{b}')
#(c)
b = a.reshape(2,3)
print(f'(c):{b}')
#(d)
b = np.random.permutation(a)
print(f'(d):{b}')
#(e)
b = a[[0,3,4]]
print(f'(e):{b}')
print(f'b是a的copy，a不變，a陣列為{a}')
```
